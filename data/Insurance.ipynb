```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint

# Create an instance of MongoClient
mongo = MongoClient(port=27017)

# confirm that our new database was created
# List the databases
print(mongo.list_database_names())


# assign the uk_food database to a variable name
db = mongo['Insurance']

# review the collections in our new database
# List the collections in the uk_food database
print(db.list_collection_names())


# review the collections in our new database
# Retrieve and display one document from the establishments collection
collection = db["insurance"]
document = collection.find_one()
pprint(document)




# assign the collection to a variable
establishments = db['insurance']
```

```python
# Retrieve unique values for the "Status" field
status_values = collection.distinct('Status')

# Print the unique status values
print(status_values)
```

```python
# Retrieve unique values for the "Company" field
company_values = collection.distinct('Company')

# Print the unique company values
print(company_values)
```

```python
# Perform aggregation to count documents per company
pipeline = [
    {
        '$group': {
            '_id': '$Company',
            'count': {'$sum': 1}
        }
    },
    {
        '$sort': {'count': -1}
    }
]
result = collection.aggregate(pipeline)

# Print the count of documents per company in descending order
for entry in result:
    print(f"Company: {entry['_id']}, Count: {entry['count']}")
```

```python

pipeline = [
    {
        '$group': {
            '_id': '$Company',
            'count': {'$sum': 1}
        }
    },
    {
        '$match': {
            'count': {'$lt': 500}
        }
    }
]
result = collection.aggregate(pipeline)

# Drop insurance companies with less than 500 documents
for entry in result:
    company_name = entry['_id']
    collection.delete_many({'Company': company_name})
    print(f"Dropped company: {company_name}")
```

```python
# Retrieve unique values for the "Coverage" field
coverage_values = collection.distinct('Coverage')

# Print the unique coverage values
print(coverage_values)
```

```python


# Retrieve unique values for the "SubCoverage" field
subcoverage_values = collection.distinct('SubCoverage')

# Print the unique subcoverage values
print(subcoverage_values)



```

```python
# Group data by "Coverage" field and calculate count
pipeline = [
    {
        '$group': {
            '_id': '$Coverage',
            'count': {'$sum': 1}
        }
    }
]
result = collection.aggregate(pipeline)

# Print the grouped data
for entry in result:
    print(f"Coverage: {entry['_id']}, Count: {entry['count']}")
```
